### Step 9: Converting Multiclass to Binary Class 

```python
df_bin = derma_df.reset_index(drop=True)
df_bin.loc[df_bin['CLASS_CODE'] <= 3, 'CLASS_CODE'] = 1
df_bin.loc[df_bin['CLASS_CODE'] > 3, 'CLASS_CODE'] = 2
df_bin
```

### Step 10: Splitting the Data

```python
# split the data into validation and training set
train_df, test_df = train_test_split(derma_df, test_size=0.3)

# to reduce repetition in later code, create variables to represent the columns
# that are our predictors and target
target = 'CLASS_CODE'
predictor = list(df_bin.columns)
predictor.remove(target)

X = df_bin[predictor]
y = df_bin[target]
```

```python
df_bin['CLASS_CODE'].value_counts()
```

#### 
As we could see that there is data imbalance in the target variable, we are addressing this issue using over sampling tehcnique.

### Step 11: Addressing the Data Imbalance

```python
oversample = RandomOverSampler(sampling_strategy='minority')

X_over, y_over = oversample.fit_resample(X, y)

print(Counter(y_over))
```

### Step 12: Save the Data

```python
train_X = train_df.drop(columns=['CLASS_CODE'])
train_y = train_df.CLASS_CODE
test_X= test_df.drop(columns=['CLASS_CODE'])
test_y = test_df.CLASS_CODE 

train_df.to_csv('disease_train_df.csv', index=False)
train_X.to_csv('disease_train_X.csv', index=False)
train_y.to_csv('disease_train_y.csv', index=False)
test_df.to_csv('disease_test_df.csv', index=False)
test_X.to_csv('disease_test_X.csv', index=False)
test_y.to_csv('disease_test_y.csv', index=False)
```
